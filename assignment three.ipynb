```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

def plot_cubic_spline(control_pts, num_curve_points=500):
    """
    Draws control points, control polygon, and natural cubic spline.

    Parameters:
    - control_pts: list of (x, y) tuples, must be in ascending x order
    - num_curve_points: resolution of the spline curve
    """
    # Unzip and convert to arrays
    xs, ys = zip(*control_pts)
    xs = np.array(xs)
    ys = np.array(ys)

    cs = CubicSpline(xs, ys, bc_type='natural')

    x_fine = np.linspace(xs.min(), xs.max(), num_curve_points)
    y_fine = cs(x_fine)


    plt.figure(figsize=(8, 5))
    # (dashed red) with control points
    plt.plot(xs, ys, 'ro--', label='Control Polygon')
    # Spline curve (solid blue)
    plt.plot(x_fine, y_fine, 'b-', linewidth=2, label='Natural Cubic Spline')

    plt.title('Natural Cubic Spline Interpolation')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```

```python
if __name__ == "__main__":
    control_pts = [
        (0.0,  0.0),
        (0.5,  0.8),
        (1.0,  1.0),
        (1.5,  0.3),
        (2.0, -0.4),
        (2.5, -1.0)
    ]

    plot_cubic_spline(control_pts)
```
